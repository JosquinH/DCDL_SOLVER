```python
import numpy as np
import time
```

```python
def create_game(game_tuile=[],resultat=0):

    if game_tuile == [] or resultat == 0:

        tab_tuile = [2,100,4,9,6,8,1,25,4,3,75,50,7,10,9,5,2,7,5,6,8,3,1]

        game_tuile = []

        for i in range(6):
            indice = np.random.randint(0,len(tab_tuile))
            game_tuile.append(tab_tuile[indice])
            del tab_tuile[indice]
            np.random.shuffle(tab_tuile)

        resultat  = np.random.randint(100,1000)

    print('Tuiles : %d, %d, %d, %d, %d, %d \n'%(game_tuile[0],game_tuile[1],game_tuile[2],game_tuile[3],game_tuile[4],game_tuile[5]))
    print('À trouver : %d\n'%resultat)
    print('--------------------------------------------\n')

    t0 = time.time()

    min_epsi,res = compute_compte_est_bon(game_tuile,resultat)

    t1 = time.time() - t0

    print('Finit en %f secondes\n'%t1)

    if min_epsi == 0:
        print('Le Compte Est Bon !!!\n')
    else:
        final_result = res[len(res)-1][3]
        print('Valeur la plus proche : %d\n'%final_result)

    print('--------------------------------------------\n')

    for op in res:
        print('-> %d %s %d = %d \n'%(op[1],op[0],op[2],op[3]))



def all_operation(tuile_list):

    all_op = []

    for i in range(len(tuile_list)):

        x1 = tuile_list[i]

        for j in range(i+1,len(tuile_list)):

            x2 = tuile_list[j]

            if x1 != 0 and x2 != 0:
                all_op.append(('+',x1,x2,x1+x2,(i,j)))

                if x1 > x2:
                    all_op.append(('-',x1,x2,x1-x2,(i,j)))
                elif x2 > x1:
                    all_op.append(('-',x2,x1,x2-x1,(i,j)))

                if x1 != 1 and x2 != 1:
                  all_op.append(('*',x1,x2,x1*x2,(i,j)))

                  if x1 >= x2 and x1 % x2 == 0:
                      all_op.append(('/',x1,x2,int(x1/x2),(i,j)))
                  elif x2 >= x1 and x2 % x1 == 0:
                      all_op.append(('/',x2,x1,int(x2/x1),(i,j)))

    return all_op


def compute_compte_est_bon(tuile_list,res):

    #Initialisation

    operation = all_operation(tuile_list)

    min_epsi = res

    result = []

    OPEN = []

    for op in operation:
        t = tuile_list.copy()

        del t[op[4][1]]
        del t[op[4][0]]

        if len(t) > 0 :
            t.append(op[3])
            OPEN.append([t,[op[0:4]]])

        if np.abs(res - op[3]) < min_epsi:
            min_epsi = np.abs(res - op[3])
            result = [op[0:4]]

    #Recherche de solution

    while OPEN != [] and min_epsi != 0:
        current_op = OPEN.pop(0)
        operation = all_operation(current_op[0])

        for op in operation:
            t = current_op[0].copy()
            del t[op[4][1]]
            del t[op[4][0]]

            if len(t) > 0 :
                t.append(op[3])
                OPEN.append([t,current_op[1] + [op[0:4]]])

            if np.abs(res - op[3]) < min_epsi:
                min_epsi = np.abs(res - op[3])
                result = current_op[1] + [op[0:4]]

    return min_epsi, result
```
